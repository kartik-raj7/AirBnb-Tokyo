In japan people renting out place need to obtain minpaku license which have the following conditions:-

**Maximum Rental Days:**

Properties can be rented out for up to 180 days per year under normal conditions. If the owner wants to rent the property for more than 180 days, they must obtain special permission or be in a special zone where longer stays are allowed.

**Registration with the Government:**

Owners must register their property with the local government. This includes providing information about the property and registering as a short-term rental operator.

The property will be assigned a license number (e.g., M130000057), which must be displayed in the listing.

**Reporting Guest Information:**

Hosts are required to report guest information to the government, including names and dates of stay. This is typically done for each guest and helps with tracking short-term rental activity.

**Safety Standards:**

The property must meet certain safety standards to be eligible for a license:

Smoke Alarm: A working smoke alarm is required in the property.

Fire Extinguisher: A fire extinguisher must be available and easily accessible.

Escape Routes: Clear evacuation routes and safety plans must be in place in case of an emergency.


The Listings above have availability of more than 180 days apart from that lack the basic amenity of smoke alarm suggesting they had one when obtaining license but havent really maintained it thereafter which jeopardizes the safety of the tourist who come to stay. The Japanese Govt should look into these places.

```python
import folium
import geopandas as gpd
import pandas as pd
from branca.colormap import LinearColormap
from fuzzywuzzy import process


adam = gpd.read_file("neighbourhoods.geojson") 

def clean_neighbourhood_name(name):
    
    name = name.lower().strip()
    name = name.replace("-", " ")

    suffixes = ['ku', 'shi', 'machi', 'city', 'ward', '区', '市', 'town']
    for suffix in suffixes:
        if name.endswith(suffix):
            name = name.replace(suffix, '').strip()

    name = name.replace(",", "").replace("japan", "").replace("tōkyō-to", "").strip()

    return name

adam['cleaned_neighbourhood'] = adam['neighbourhood'].apply(lambda x: clean_neighbourhood_name(x))
avg_price_per_neighbourhood['cleaned_neighbourhood'] = avg_price_per_neighbourhood['neighbourhood'].apply(lambda x: clean_neighbourhood_name(x))
def get_geometry(neighbourhood):
    match = process.extractOne(neighbourhood, adam['cleaned_neighbourhood'])
    
    if match: 
        matched_neighbourhood = match[0] 
        matched_row = adam[adam['cleaned_neighbourhood'] == matched_neighbourhood]  # Find the row
        
        if not matched_row.empty:
        
            return matched_row['geometry'].values[0]
    return None
avg_price_per_neighbourhood['coordinates'] = avg_price_per_neighbourhood['cleaned_neighbourhood'].apply(get_geometry)


color_scale = LinearColormap(['yellow', 'red'], vmin=min(avg_price_per_neighbourhood['price'].dropna()), vmax=max(avg_price_per_neighbourhood['price'].dropna()))


def get_color(feature):
    value = feature['properties'].get('price', None)
    
    if value is None:
        return 'gray'  
    
    return color_scale(value)

gdf = gpd.GeoDataFrame(avg_price_per_neighbourhood, geometry='coordinates')
map3 = folium.Map(location=[35.6895, 139.6917], zoom_start=11)
gdf.to_file("avg_price_neighbourhood.geojson", driver="GeoJSON")

folium.GeoJson(
    data=gdf.to_json(), 
    name='Neighborhoods',
    tooltip=folium.features.GeoJsonTooltip(
        fields=['neighbourhood', 'price','avg_review_score'],
        aliases=['Neighborhood', 'Average Price','Average Ratings'],
        labels=True,
        sticky=False
    ),
    style_function=lambda feature: {
        'fillColor': get_color(feature),
        'color': 'black',
        'weight': 1,
        'dashArray': '5, 5',
        'fillOpacity': 0.5
    },
    highlight_function=lambda feature: {
        'weight': 3,
        'fillColor': get_color(feature),
        'fillOpacity': 0.8
    }
).add_to(map3)

map3
```

```python
from sqlalchemy import text

query = text("""
    SELECT
        id,
        name,
        host_id,
        host_name,
        number_of_reviews,
        review_scores_rating,
        price,
        latitude,
        longitude,
        last_review
    FROM listings_cleaned
    WHERE number_of_reviews <= 10
      AND review_scores_rating <= 3.0
      AND price IS NOT NULL
       AND last_review < CURRENT_DATE - INTERVAL 1 YEAR
""")


with engine.connect() as conn:
    df = pd.read_sql(query, conn)

import re

def clean_price(price_str):
    if isinstance(price_str, str):
        cleaned = re.sub(r'[^\d.]', '', price_str)
        try:
            return float(cleaned)
        except ValueError:
            return None
    return price_str

df['price_num'] = df['price'].apply(clean_price)

df = df.dropna(subset=['price_num'])

high_price_low_review = df[df['price_num'] > 2000].sort_values(by='price_num', ascending=False)

display(high_price_low_review)
```

```python
host_summary = (
    high_price_low_review
    .groupby(['host_id', 'host_name'])
    .agg(
        num_listings=('id', 'count'),
        avg_price=('price_num', 'mean'),
        avg_number_of_reviews=('number_of_reviews', 'mean'),
        max_price=('price_num', 'max'),
        min_price=('price_num', 'min'),
        total_price=('price_num', 'sum')
    )
    .reset_index()
    .sort_values(by='num_listings', ascending=False)
)

display(host_summary.head(20))
```
